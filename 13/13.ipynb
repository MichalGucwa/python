Read any raster image using the PIL library. Print some attributes (mode, size, format). <br>Send me the script, the results, and the thumbnail with size=(128,128). 

```python
from PIL import Image

#Author of image: © kuritafsheen—RooM/Getty Images
#article
#https://www.britannica.com/story/whats-the-difference-between-a-frog-and-a-toad
#Image link
#https://cdn.britannica.com/84/206384-050-00698723/Javan-gliding-tree-frog.jpg?q=60
with Image.open("frog.jpeg") as im:
    print("\nAtributes")
    print("Mode: %s" % (im.mode))
    print("Size (width, height): ", im.size)
    print("Format:", im.format)
    im_org=im.copy()
    im.thumbnail([128,128])
    im_thumbnail=im
    im_resize=im_org.resize([128,128])
    print("thumnail_size (width, height):", im.size)
    im_resize.save('thumbnail.jpeg')
```
