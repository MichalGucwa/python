# EXERCISE 9.1 (PLOT)

Plot functions sin(x), cos(x), and exp(-x) in a range [0,10]. Colors are red, green, and blue, respectively. Lines are solid, dashed, and dotted, respectively. Add a legend. 

```python
import numpy as np
from matplotlib import pyplot as plt

X=np.arange(0,10,0.01)
Y_sin=np.sin(X)
Y_cos=np.cos(X)
y_exp=np.exp(-X)

ax=plt.subplot()
ax.plot(X,Y_sin, label='sin(x)', color='red', linestyle='-')
ax.plot(X,Y_cos, label='cos(x)', color='green', linestyle='-.')
ax.plot(X,y_exp, label='exp(-x)', color='blue', linestyle=':')
ax.legend()
```

# EXERCISE 9.2 (SCATTER)

Generate n=100 random points in a unit square [0,1]x[0,1]. Points are green if the distance from (0,0) is less then 1; they are red otherwise. The marker area of a point (x,y) should be proportional to x+y. 

```python
n=100
X=[np.random.rand() for _ in range(n)]
Y=[np.random.rand() for _ in range(n)]
vector_len=[(x**2+y**2)**(0.5) for x,y in zip(X,Y)]
colors=['green' if val<=1 else 'red' for val in vector_len]
marker_area=np.array([x+y for x,y in zip(X,Y)])*10

ax=plt.subplot()
ax.scatter(X,Y, s=marker_area, color=colors)
ax.legend()
```
